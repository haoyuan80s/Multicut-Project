```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

import pickle
import numpy as np
import math
#print(data.keys())
def fetch(res,key):
    return np.mean(list(map(lambda x: x[key],res)))
```

```python
data = pickle.load(open("data/rand/rand_data.pkl",'rb'))

keys = sorted(list(data.keys()))
x = np.array(keys)
outy = []
outz = []
for n in keys:
    res = data[n]
    outy.append(fetch(res,'obj_LP'))
    outz.append(fetch(res,'obj_RG'))
y = np.array(outy)
z = np.array(outz)
fig, ax = plt.subplots()
plt.title('Random Graph') 
ax.plot(x,y, linewidth=3,label='Linear program')
ax.plot(x,z, linewidth=3,label='Region growing')
#plt.plot(x,y*4*math.log(4))
legend = ax.legend(loc='upper center')
plt.xlabel('Number of nodes')
plt.ylabel('Objective value')
#plt.show()
plt.savefig('Random Graph Value.pdf')
```

```python
data = pickle.load(open("data/grid/small_n.pickle",'rb'))
keys = sorted(list(data.keys()))
x = np.array(keys)
outy = []
outz = []
outw = []
for n in keys:
    res = data[n]
    outy.append(fetch(res,'obj_IP'))
    outz.append(fetch(res,'obj_RG'))
    outw.append(fetch(res,'obj_LP'))
y = np.array(outy)
z = np.array(outz)
w = np.array(outw)
fig, ax = plt.subplots()
plt.title('IP vs LP vs Region Growing') 
ax.plot(x,y,linewidth=3, label='IP OPT')
ax.plot(x,z,linewidth=3, label='Region growing')
ax.plot(x,w,linewidth=3, label='LP relax')
#plt.plot(x,y*4*math.log(4))
legend = ax.legend(loc='upper center')
plt.xlabel('Number of nodes')
plt.ylabel('Objective value')
#plt.show()
plt.savefig('IP vs LP vs Region Growing.pdf')
```
